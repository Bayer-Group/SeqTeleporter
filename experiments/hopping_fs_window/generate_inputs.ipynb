```python
from proseqteleporter.random_sample_generator.random_sample_generator import random_sample_generator

AMINO_ACIDS = 'ACDEFGHIKLMNPQRSTVWY'
```

```python
inputs = dict(SEQUENCE=None,
              FIX_DNA_SEQUENCE="",
              MUTATIONS=None,
              LINKED_MUTATIONS=None,
              CUT_NUMBER_RANGE=(0,6),
              MIN_AA_LENGTH=6,
              MAX_COST=10000000,
              MAX_LENGTH_UNEVENNESS=49,
              MIN_LIGATION_FIDELITY=0.98,
              SATISFACTION_LIGATION_FIDELITY=0.98,
              FIDELITY_DATA="FileS01_T4_01h_25C.xlsx",
              HOST="c_griseus",
              FUSION_SITES_USED_BY_BACKBONE=('CTTG', 'TAAT'),
              DNA_5_PRIME="AATTTGGTCTCTCC",
              DNA_3_PRIME="TAATAGAGACCTTTAA",
              ALLOWED_CUT_POSITIONS=[],
              COST_PER_BP=0.1,
              MIN_DNA_LENGTH=1,
              MAX_DNA_LENGTH=10000000,
              MODULE_PLATE_FORMAT=96,
              ENZYME='BsaI')
```

```python
from os import path, mkdir
SCRIPT_DIR = path.dirname(path.abspath('__file__'))


for rep in range(5):
    inputs_dir = path.join(SCRIPT_DIR,f'rep{rep}')
    if not path.isdir(inputs_dir):
        mkdir(inputs_dir)
    for aa_len in range(40,80,10):
        s, mutations_1idx, linked_mutations_1dx_ = random_sample_generator(
            min_aa_length=aa_len, max_aa_length=aa_len,
            min_number_of_positions=3, max_number_of_positions=aa_len // inputs['MIN_AA_LENGTH'] - 1,
            min_variations_per_position=1, max_variations_per_position=19,
            max_positions_per_linked_mutation_set=0,
            max_number_of_mutation_linked_mutation_sets=0
        )
        inputs.update({'SEQUENCE':s, 'MUTATIONS':mutations_1idx})
        with open(path.join(inputs_dir,f'{aa_len}aa.txt'), 'w') as f:
            for k, v in inputs.items():
                f.write(f'{k}={v}\n')
```
