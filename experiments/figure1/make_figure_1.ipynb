```python
Xdf = pd.DataFrame(
    list(range(100,1000,20))+list(range(1000,500,-10))+list(range(500,-500,-50))+list(range(-500,-580,-1)),
    index=list(range(0,195)),columns=['Val']
)
Ydf = pd.DataFrame(
    list(range(220,300,2))+list(range(300,3000,60))+list(range(3000,2000,-40))+list(range(2000,2500,10))+list(range(2400,600,-60)),
    index=list(range(0,190)),columns=['Val']
)
```

```python
# fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax = plt.figure(figsize=(6, 6)).add_subplot(projection='3d')

X = Xdf.index
Y = Ydf.index
Z = pd.DataFrame(index=X, columns=Y)
for idx, i in zip(X, Xdf['Val']):
       for idx2, i2 in zip(Y, Ydf['Val']):
              Z.loc[idx,idx2] = i*i2
X, Y = np.meshgrid(X, Y, sparse=False)
# Plot the surface.
surf = ax.plot_surface(X, Y, Z.transpose(), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
# ax.set_zlim(-4.11, 4.11)
ax.zaxis.set_major_locator(LinearLocator(50))
# A StrMethodFormatter is used automatically
# ax.zaxis.set_major_formatter('{x:.01f}')
# ax.xaxis.set_major_formatter('{x:.0f}')
# ax.yaxis.set_major_formatter('{x:.0f}')
ax.xaxis.set_major_locator(LinearLocator(50))
ax.yaxis.set_major_locator(LinearLocator(50))
ax.set(
       xticklabels=[],
       yticklabels=[],
       zticklabels=[],
       # xlabel='sequence space',
       # ylabel='sequence space',
       # zlabel='fitness'
       )
# ax.zaxis.labelpad=5
# zLabel = ax.set_zlabel('Z zzzz zzz (z)', linespacing=0.4)
# ax.zaxis._axinfo['juggled'] = (-1, 0, 0)
# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
