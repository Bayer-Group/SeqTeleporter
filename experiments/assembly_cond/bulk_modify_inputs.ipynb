```python
import os
from os.path import join
from os import path, mkdir
import re


SCRIPT_DIR = path.dirname(path.abspath('__file__'))


def modify_input(old_input_file, new_input_file, param, new_value):

    with open(new_input_file, 'w') as new_file:
        with open(old_input_file,'r') as old_file:
            for line in old_file:
                if re.search(param,line):
                    print(line)
                    new_file.write(f'{param}={new_value}\n')
                else:
                    new_file.write(line)
```

```python
def list_input_files(input_dirs):
    sel_input_files = []
    for input_dir in input_dirs:
        for item in os.listdir(join(SCRIPT_DIR,  input_dir)):
            full_path = join(SCRIPT_DIR,  input_dir, item)
            if os.path.isfile(full_path) and re.search(f'[.]txt$',item):
                sel_input_files.append(full_path)
    return sel_input_files
```

```python
input_files = list_input_files(["fix_cplx_diff_muts_inputs_T4_01h_25C"])
for fidelity_data in ["FileS08_T7_18h_37C.xlsx","FileS06_T7_18h_25C.xlsx"]:
    output_dir = join(SCRIPT_DIR,f'fix_cplx_diff_muts_inputs{re.sub("FileS[0-9][0-9]|[.xlsx]","",fidelity_data)}')
    if not path.isdir(output_dir):
        mkdir(output_dir)
    for input_file in input_files:
        input_file_name = re.sub("^.*fix_cplx_diff_muts_inputs_T4_01h_25C\\\\","",input_file)
        # print(input_file_name)
        modify_input(old_input_file=input_file,
                     new_input_file=join(output_dir, input_file_name),
                     param="FIDELITY_DATA",
                     new_value=fidelity_data)
```
